```python
import lyricsgenius
import re
import pandas as pd

with open('geniustoken.txt', 'r') as file:
    token = file.read().strip()
    file.close()

genius = lyricsgenius.Genius(token)

def get_first_artist(artist):
    pattern = '[^,|&]*'
    return re.match(pattern, artist).group()

def get_lyrics(artist_song_list):
    return [[artist, song, genius.search_song(song, get_first_artist(artist), get_full_info=False).lyrics] for
            artist, song in artist_song_list]

data = pd.read_csv("kerim_songs.csv")
n_df= pd.DataFrame()
n_df["names"] = ["ad","ads","gd"]
n_df["ages"] = [1,2,3]
print(n_df.head())


artist_list = []
for i in range(data["artist"].shape[0]):
    lst = [data["artist"][i], data["song"][i]]
    artist_list.append(lst)

#artist_list = [['Sharon Van Etten', 'Seventeen'], ['P Diddy, Usher & Loon', 'I Need A Girl'], ['Jay-Z & Kanye West', 'Niggas in Paris']]

lyrics = get_lyrics(artist_list)
#print(lyrics)

def remove_brackets(lyrics_list):
    pattern = '\[.*\]'
    for song in lyrics:
        song[2] = re.sub(pattern, '', song[2])

remove_brackets(lyrics)
print(lyrics[0])

data["lyrics"] = lyrics

print(data.head())

```
