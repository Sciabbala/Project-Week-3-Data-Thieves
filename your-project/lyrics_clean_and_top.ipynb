```python
import operator
import pandas as pd
import re, string

def clean_lyrics(str):
    res = re.sub('[%s]' % re.escape(string.punctuation), ' ', str)
    res = res.replace("\n", " ")
    return res

common = ["clap", "oh", "hey", "not", "ahh", "eheu", "didn" ,"huh", "ooh" ,"wouldn", "bout", "that", "ayy", "yeah", "don", "we", "dont", "youre", "but", "no", "yes", "if", "to", "im", 'i' ,'how', 'where', 'what', 'when', 'why', 'who', 'very', 'really', 'quite', 'somewhat', 'rather', 'too', 'may', 'might', 'can', 'could', 'will', 'would', 'shall', 'should', 'be', 'is', 'am', 'are', 'have', 'has', 'do', 'does', 'did', 'get', 'got', 'was', 'were', 'she', 'they', 'he', 'it', 'him', 'her', 'you', 'me', 'anybody', 'somebody', 'someone', 'anyone', 'in', 'of', 'between', 'on', 'with', 'by', 'at', 'without', 'through', 'over', 'across', 'around', 'into', 'within ', 'and', 'but', 'for', 'yet', 'neither', 'or', 'so', 'when', 'although', 'however', 'as', 'because', 'before ,some', 'both', 'most', 'many', 'any', 'much', 'enough', 'several', 'none', 'all', 'my', 'your', 'their', 'our', 'ours', 'whose', 'his', 'hers', 'its', 'which  that', 'this', 'those', 'these', 'a', 'an', 'the']
def find_top(str, num):
    word_counter = {}
    for word in str.split(" "):  # split in every space.
        if word != '\r\n' and word not in common and len(word) > 2: # CHECK COMMON WORDS HERE
            if word not in word_counter:  # if 'word' not in word_counter, add it, and set value to 1
                word_counter[word] = 1
            else:
                word_counter[word] += 1  # if 'word' already in word_counter, increment it by 1

    sorted_d = sorted(word_counter.items(), key=operator.itemgetter(1))

    return sorted_d[-num:]

data = pd.read_csv("kerim_onlylyrics.csv")


clean_lyrics_list = []

for i in range(data.shape[0]):
    ly = data["lyrics"][i]
    clean = clean_lyrics(ly)
    clean_lyrics_list.append(clean.lower())

data["clean_lyrics"] = clean_lyrics_list

top_words_list = []
for i in range(data.shape[0]):
    top_words_list.append(find_top(data["clean_lyrics"][i],4))
data["top"] = top_words_list
data
```
